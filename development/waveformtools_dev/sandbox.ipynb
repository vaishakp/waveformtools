```python
def progressbar(present_count,total_counts, normalize = 'no'):
        ''' Display the progress bar to std out from present_count and total_count.
        
        -----------
        **Input:** 
            present_count (int) : The present count state. 
            total_counts(int)   : The final state.
        

        **Returns:** 
            1
            The progress bar is printed to stdout.
        
        '''
        if normalize=='yes':
            present_count = int(100.0*present_count/total_counts)*10.0
            total_counts = 100*10
            
        #present_count = comm.gather(count,root=rank)
        final_progress=(int(total_counts/10)+1)
        present_stage=int(present_count/10)
        print(present_count)
        print(present_count%10)
        #print(present_stage)
        if present_stage!=0:
            #print('not 0')
            present_semi_progress= present_count%10
            #print(present_semi_progress)
        else:
            #print('0')
            present_semi_progress=present_count
            #print(present_semi_progress)
        #print(present_semi_progress)
        offset=final_progress-present_stage
        sys.stdout.write('\r' + "Progress:|" + '#'*present_stage +'%s'%present_semi_progress+' '*offset +'|')
        sys.stdout.flush()
        return 1
```

```python
N=100
for index in range(0,N):
    progressbar(index, N)
```

```python
def progressbar2(present_count,total_counts, normalize = 'yes'):
        ''' Display the progress bar to std out from present_count and total_count.
        
        -----------
        **Input:** 
            present_count (int) : The present count state. 
            total_counts(int)   : The final state.
        

        **Returns:** 
            1
            The progress bar is printed to stdout.
        
        '''
        
            
        if normalize=='yes':
            final_progress=98
            normalized_total_counts=final_progress *10
            present_count = int(normalized_total_counts*present_count/total_counts)
            total_counts = normalized_total_counts
            
        #present_count = comm.gather(count,root=rank)
        else:
             final_progress=(int(total_counts/10))
        
        present_stage=int(present_count/10)
        #print(present_stage)
        #print(total_counts)
        if present_stage!=0:
            if present_count==total_counts:
                present_semi_progress='#'
            else:
                present_semi_progress=present_count%10
        else:
            present_semi_progress=present_count
        offset=final_progress-present_stage
        pc = 100*present_count/total_counts
        
        sys.stdout.write('\r' + "Progress:|" + '#'*present_stage +'%s'%present_semi_progress+' '*(offset) +'|'+'%.f%%'%pc)
        sys.stdout.flush()
        return 1
```
